```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. 1s ---
# 1s r 
#  ^2 
def probability_density_1s(x, y, z):
    """
    1s
    ()
    """
    #  r  ( a0)
    r = np.sqrt(x**2 + y**2 + z**2)
    #  (e^(-2r) )
    # 
    return np.exp(-2 * r)

# --- 2. 3 ---
# 
limit = 4.0
resolution = 50  #  ()

# x, y, z 
x = np.linspace(-limit, limit, resolution)
y = np.linspace(-limit, limit, resolution)
z = np.linspace(-limit, limit, resolution)

# 3D
X, Y, Z = np.meshgrid(x, y, z)

# --- 3.  ---
# 
P = probability_density_1s(X, Y, Z)

# --- 4.  (/) ---
# 
# 
threshold = np.max(P) * 0.01  # 10%

# 
# 
mask = P > threshold

# 
# X[mask]  P > threshold  X 
x_plot = X[mask]
y_plot = Y[mask]
z_plot = Z[mask]

# --- 5. Matplotlib ---
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 
# s=2 , alpha=0.1 
ax.scatter(x_plot, y_plot, z_plot, c=z_plot, cmap='coolwarm', s=2, alpha=0.1, marker='o')

# 
ax.set_xlabel('X (a0)')
ax.set_ylabel('Y (a0)')
ax.set_zlabel('Z (a0)')
ax.set_title('1s Orbital (Electron Probability Density)')

# 
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# 
plt.show()
```

## 

### Subtask:
1s2s


**Reasoning**:
Define the approximate wave functions for the 1s and 2s orbitals of a Lithium atom as specified in the instructions.



```python
#  ()
Zeff_1s = 2.69
Zeff_2s = 1.59

def psi_1s_lithium(x, y, z):
    """
    1s
    """
    r = np.sqrt(x**2 + y**2 + z**2)
    # 1sZeff
    return np.exp(-Zeff_1s * r)

def psi_2s_lithium(x, y, z):
    """
    2s
    """
    r = np.sqrt(x**2 + y**2 + z**2)
    # 2s
    return (1 - Zeff_2s * r / 2) * np.exp(-Zeff_2s * r / 2)

print("Defined psi_1s_lithium and psi_2s_lithium functions.")
```

## 

### Subtask:



**Reasoning**:
Calculate the probability densities for the 1s and 2s orbitals using the previously defined wave functions and the 3D grid.



```python
# 1s
psi_1s_vals = psi_1s_lithium(X, Y, Z)

# 1s ()
P_1s = psi_1s_vals**2

# 2s
psi_2s_vals = psi_2s_lithium(X, Y, Z)

# 2s ()
P_2s = psi_2s_vals**2

print("Calculated probability densities for 1s and 2s orbitals.")
```

## 3d

### Subtask:
3D1s2s


**Reasoning**:
Prepare the data for plotting by applying thresholds and extracting coordinates for both 1s and 2s orbitals.



```python
# 1s
threshold_1s = np.max(P_1s) * 0.01

# 1s
mask_1s = P_1s > threshold_1s

# 1s
x_plot_1s = X[mask_1s]
y_plot_1s = Y[mask_1s]
z_plot_1s = Z[mask_1s]

# 2s
# 1s
threshold_2s = np.max(P_2s) * 0.05

# 2s
mask_2s = P_2s > threshold_2s

# 2s
x_plot_2s = X[mask_2s]
y_plot_2s = Y[mask_2s]
z_plot_2s = Z[mask_2s]

print("Prepared plotting data for 1s and 2s orbitals.")
```

## 

### Subtask:
1s2s3D


**Reasoning**:
Visualize the 1s and 2s orbitals using the prepared plotting data by creating a 3D scatter plot for each orbital within the same figure, setting appropriate labels and title.



```python
# --- 5. Matplotlib ---
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# 1s
# 
ax.scatter(x_plot_1s, y_plot_1s, z_plot_1s, c=z_plot_1s, cmap='Blues', s=5, alpha=0.2, marker='o', label='1s Orbital')

# 2s
# 
ax.scatter(x_plot_2s, y_plot_2s, z_plot_2s, c=z_plot_2s, cmap='Reds', s=5, alpha=0.2, marker='o', label='2s Orbital')


# 
ax.set_xlabel('X (a0)')
ax.set_ylabel('Y (a0)')
ax.set_zlabel('Z (a0)')
ax.set_title('Lithium Atom Orbitals (1s and 2s)')

# 
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# 
ax.legend()

# 
plt.show()
```
