```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. 波動関数の定義（簡略化された1s軌道） ---
# 1s軌道の波動関数（Ψ）は、原子核からの距離 r に対して指数関数的に減少します。
# 電子の存在確率密度は Ψ^2 に比例します。
def probability_density_1s(x, y, z):
    """
    1s軌道の電子存在確率密度を近似計算する関数。
    (厳密には水素原子の波動関数の二乗に比例)
    """
    # 原子核からの距離 r を計算 (単位はボーア半径 a0)
    r = np.sqrt(x**2 + y**2 + z**2)
    # 確率密度 (e^(-2r) に比例)
    # 距離が離れるほど急激に減少する球対称な形
    return np.exp(-2 * r)

# --- 2. 3次元グリッドの作成 ---
# 空間の範囲を設定
limit = 4.0
resolution = 50  # 解像度 (高くすると滑らかになるが計算時間が増加)

# x, y, z 軸のグリッド点を作成
x = np.linspace(-limit, limit, resolution)
y = np.linspace(-limit, limit, resolution)
z = np.linspace(-limit, limit, resolution)

# 3Dメッシュグリッドの作成
X, Y, Z = np.meshgrid(x, y, z)

# --- 3. 確率密度の計算 ---
# 各点での確率密度を計算
P = probability_density_1s(X, Y, Z)

# --- 4. 可視化 (等確率面/アイソサーフェス) ---
# 電子の存在確率が高い領域の境界値を設定
# この値を変更することで、軌道の輪郭のサイズが変わります
threshold = np.max(P) * 0.01  # 最大確率の10%を境界とする

# 描画用のデータを作成
# 境界値以上の確率密度を持つ点のみをプロットするためのマスク
mask = P > threshold

# 可視化に必要な座標を抽出
# X[mask] は P > threshold の条件を満たす X の要素
x_plot = X[mask]
y_plot = Y[mask]
z_plot = Z[mask]

# --- 5. Matplotlibでプロット ---
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 散布図としてプロット
# s=2 は点のサイズ, alpha=0.1 は透明度
ax.scatter(x_plot, y_plot, z_plot, c=z_plot, cmap='coolwarm', s=2, alpha=0.1, marker='o')

# 軸ラベルの設定
ax.set_xlabel('X (a0)')
ax.set_ylabel('Y (a0)')
ax.set_zlabel('Z (a0)')
ax.set_title('1s Orbital (Electron Probability Density)')

# 軸の目盛りを非表示にして原子核を中心に表現
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# プロット表示
plt.show()
```

## リチウム原子の波動関数の定義

### Subtask:
リチウム原子の1s軌道と2s軌道の波動関数を定義します。これらの関数は、水素原子のものよりも複雑になります。


**Reasoning**:
Define the approximate wave functions for the 1s and 2s orbitals of a Lithium atom as specified in the instructions.



```python
# 有効核電荷の定数 (文献値に基づく近似)
Zeff_1s = 2.69
Zeff_2s = 1.59

def psi_1s_lithium(x, y, z):
    """
    リチウム原子の1s軌道の波動関数を近似計算する関数。
    """
    r = np.sqrt(x**2 + y**2 + z**2)
    # 水素原子の1s波動関数にZeffを適用した形に比例
    return np.exp(-Zeff_1s * r)

def psi_2s_lithium(x, y, z):
    """
    リチウム原子の2s軌道の波動関数を近似計算する関数。
    """
    r = np.sqrt(x**2 + y**2 + z**2)
    # 節を持つ2s軌道の形に比例
    return (1 - Zeff_2s * r / 2) * np.exp(-Zeff_2s * r / 2)

print("Defined psi_1s_lithium and psi_2s_lithium functions.")
```

## 確率密度の計算

### Subtask:
各軌道の波動関数から確率密度を計算します。


**Reasoning**:
Calculate the probability densities for the 1s and 2s orbitals using the previously defined wave functions and the 3D grid.



```python
# 1s軌道の波動関数を計算
psi_1s_vals = psi_1s_lithium(X, Y, Z)

# 1s軌道の確率密度を計算 (波動関数の二乗)
P_1s = psi_1s_vals**2

# 2s軌道の波動関数を計算
psi_2s_vals = psi_2s_lithium(X, Y, Z)

# 2s軌道の確率密度を計算 (波動関数の二乗)
P_2s = psi_2s_vals**2

print("Calculated probability densities for 1s and 2s orbitals.")
```

## 3d空間でのプロットデータの準備

### Subtask:
計算した確率密度を3D空間にプロットするためのデータを準備します。1s軌道と2s軌道で別々に、または重ねて表示できるようにデータを整理します。


**Reasoning**:
Prepare the data for plotting by applying thresholds and extracting coordinates for both 1s and 2s orbitals.



```python
# 1s軌道の可視化用しきい値
threshold_1s = np.max(P_1s) * 0.01

# 1s軌道のマスクを作成
mask_1s = P_1s > threshold_1s

# 1s軌道のプロット用座標を抽出
x_plot_1s = X[mask_1s]
y_plot_1s = Y[mask_1s]
z_plot_1s = Z[mask_1s]

# 2s軌道の可視化用しきい値
# 1sとは異なる値に設定することも可能
threshold_2s = np.max(P_2s) * 0.05

# 2s軌道のマスクを作成
mask_2s = P_2s > threshold_2s

# 2s軌道のプロット用座標を抽出
x_plot_2s = X[mask_2s]
y_plot_2s = Y[mask_2s]
z_plot_2s = Z[mask_2s]

print("Prepared plotting data for 1s and 2s orbitals.")
```

## 電子軌道の可視化

### Subtask:
準備したデータを用いて、リチウム原子の電子軌道（1s軌道と2s軌道）を3Dで可視化します。


**Reasoning**:
Visualize the 1s and 2s orbitals using the prepared plotting data by creating a 3D scatter plot for each orbital within the same figure, setting appropriate labels and title.



```python
# --- 5. Matplotlibでプロット ---
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# 1s軌道のプロット
# 青系の色で球状にプロット
ax.scatter(x_plot_1s, y_plot_1s, z_plot_1s, c=z_plot_1s, cmap='Blues', s=5, alpha=0.2, marker='o', label='1s Orbital')

# 2s軌道のプロット
# 赤系の色で節を持つ形状をプロット
ax.scatter(x_plot_2s, y_plot_2s, z_plot_2s, c=z_plot_2s, cmap='Reds', s=5, alpha=0.2, marker='o', label='2s Orbital')


# 軸ラベルの設定
ax.set_xlabel('X (a0)')
ax.set_ylabel('Y (a0)')
ax.set_zlabel('Z (a0)')
ax.set_title('Lithium Atom Orbitals (1s and 2s)')

# 軸の目盛りを非表示にして原子核を中心に表現
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# 凡例の表示
ax.legend()

# プロット表示
plt.show()
```
